```python
# Import required libraries
import time  # Time-related functions :-Adding delays in your code
import csv  # CSV file handling
import requests  # HTTP requests :-web scraping or interacting with web 
import pandas as pd  # Data manipulation using pandas
import selenium  # Web automation using Selenium :-control a browser programmatically and  Web automation browser
from selenium import webdriver  # Webdriver for browser automation :-interacting with browser
from selenium.webdriver.edge.service import Service  # Service class for Edge webdriver :-configure the service parameters for the Edge webdriver
from selenium.webdriver.edge.options import Options  # Options class for Edge webdriver :- configure the option for  browser
from bs4 import BeautifulSoup  # HTML parsing using BeautifulSoup :-pulling data out of HTML and XML files and serching and modifing phrase tree


# Here is the code
# Function to scrape data from a given URL
def scrape_data(url):
    # Create an empty list to store dataframes
    dfs = []

    # existing code for scraping data
    driver.get(url)
    
    # Get the initial height of the webpage
    last_page_no = driver.execute_script('return document.body.scrollHeight')
    # Print the URL being scraped
    print(f"Scraping data from: {url}")

    # Scroll to the bottom of the page to load more content
    while True:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        time.sleep(3)
        new_height = driver.execute_script('return document.body.scrollHeight')
        if new_height == last_page_no:   # Break the loop if no new content is loaded
            break
        last_page_no = new_height
        
    # Use BeautifulSoup to parse the HTML of the loaded page
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    
    # Find all car information elements on the page
    cars_info = soup.find_all('div', {'class': 'NewUcExCard posR'})
    len(cars_info)  # Print the number of cars found on the page

    for cars in cars_info: # Loop through each car element and extract relevant information
        try:
            car_Name = cars.find('h3', class_='title').text
            car_price = cars.find('div', class_='Price hover').text
            car_kms = cars.find('div', class_='dotsDetails').text
            car_location = cars.find('div', class_="distanceText").text
            car_emi = cars.find('div', class_="emitext hover").text

            # Append data to the data frame
            dfs.append(pd.DataFrame({'car_Name': [car_Name], 'car_price': [car_price], 'car_kms': [car_kms],
                                     'car_location': [car_location], 'car_emi': [car_emi]}))
        except Exception as e:  # Handle any exceptions that may occur during data extraction
            print(f"Error scraping data: {e}")
            pass  # Continue to the next iteration if an exception occurs

    # Return the DataFrame with scraped data
    return pd.concat(dfs, ignore_index=True)

# Provide the path to the Edge WebDriver executable
webdriver_path = r"C:\Users\venkat naidu\Downloads\edgedriver_win64\msedgedriver.exe"

# Create an EdgeOptions instance
options = Options()  # Allows you to set various options for the Edge browser

# Add any additional options if needed
# Set the path for the Edge driver executable in the EdgeOptions :- can  where selenium lokk for web browser msedge.exe
options.binary_location = r"C:\Program Files (x86)\Microsoft\Edge\Application\msedge.exe"

# Create the webdriver instance using Service
driver = webdriver.Edge(service=Service(executable_path=webdriver_path), options=options)

try:
     # List of URLs for different locations or categories to scrape data from
    urls_to_scrape = [     # Iam Adding  more URLs here
        
        #all_india_used cars:
         "https://www.cardekho.com/used-cars+in+india",
        #ahmedabad:-
         "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+ahmedabad",
        #bangalore:-
         "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+bangalore",
        #mumbai:-
         "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+mumbai",
        #hyderabad:-
         "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+hyderabad",
        #chennai:-
         "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+chennai",
        #pune:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+pune",
        #jaipur:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+jaipur",
        #delhi:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+delhi-ncr",
        #agra:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+agra",
        #lucknow:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+lucknow",
        #gurgon:-
          "https://www.cardekho.com/used-maruti+hyundai+honda+tata+mahindra+toyota+renault+ford+volkswagen+skoda+mercedes-benz+kia+audi+bmw+mg+chevrolet+nissan+fiat+datsun+jeep+land-rover+isuzu+force+opel+reva+bentley+porsche+mitsubishi+citroen+volvo+jaguar+mini+mahindra-ssangyong+cars+0-lakh-to-2-lakh+2-lakh-to-3-lakh+3-lakh-to-5-lakh+5-lakh-to-8-lakh+8-lakh-to-10-lakh+10-lakh-to-5-crore+in+gurgaon",
       
        
         ]

    # Initialize an empty list to store DataFrames
    all_dfs = []
    
    
     # Scrape data from the current URL and store it in a DataFrame
    for url in urls_to_scrape:
        scraped_df = scrape_data(url)
        all_dfs.append(scraped_df) # Append the scraped DataFrame to the list of all DataFrames

    # Concatenate all DataFrames from different URLs into a single DataFrame
    final_df = pd.concat(all_dfs, ignore_index=True)

    # Print the final DataFrame with scraped data from all URLs
    print(final_df)

finally:
    # Close the webdriver
    driver.quit()
```
