```python
#removing boolean values and replace No EMI option@
# Assuming df is your DataFrame
df['car_emi'].fillna('No EMI option @', inplace=True)  

# Display the updated DataFrame to check 
print(df)
```

```python
# Extract years and car names
df['year'] = df['car_Name'].str.extract(r'(\d{4})')

df['car_name'] = df['car_Name'].str.replace(r'\b\d{4}\b', '', regex=True).str.strip()


# Display the updated DataFrame
print(df[['year', 'car_name']])
```

```python
df = df.drop(columns=['car_Name'])
df
```

```python
# Assuming df is your DataFrame
df.rename(columns={'year': 'car_Year'}, inplace=True)

# Display the updated DataFrame
print(df)
```

```python
# Display the current column names
print(df.columns)

# Make sure the column names are spelled correctly
# Then reorder the columns as needed
df = df[['car_name', 'car_price', 'car_kms', 'car_location', 'car_emi', 'car_Year']]

# Display the updated DataFrame
print(df)
```
